# 1) What is a color space ? 
A specification of a coordinate system and subspace within that system, where each color is represented by a single point.

# 2) What are the different color spaces ? 

RGB (Red, Green, Blue):
   RGB is an additive color space commonly used in digital displays and imaging. Each color is represented by the intensity of red, green, and blue components.

```python
from PIL import Image

image = Image.open("image.jpg")  # Open the image
rgb_image = image.convert("RGB")  # Convert to RGB color space
```

HSL (Hue, Saturation, Lightness): HSL is a color space that represents colors based on their hue, saturation, and lightness values. It is often used for color selection and manipulation.

```python
import cv2
import numpy as np

image = cv2.imread("image.jpg")  # Read the image
hsl_image = cv2.cvtColor(image, cv2.COLOR_BGR2HLS)  # Convert to HSL color space
```

Lab (CIELAB or Lab*): Lab is a device-independent color space designed to approximate human vision. It separates the lightness (L*) component from the a* and b* components representing color information.

```python
from skimage import io, color

image = io.imread("image.jpg")  # Read the image
lab_image = color.rgb2lab(image)  # Convert to Lab color space
```

Let's compare the different images in different color spaces.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('image.jpg')

# Convert to RGB
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Convert to HSL
hsl_image = cv2.cvtColor(image, cv2.COLOR_BGR2HLS)

# Convert to LAB
lab_image = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)

# Convert to YUV
yuv_image = cv2.cvtColor(image, cv2.COLOR_BGR2YUV)

# Create a figure to display the images
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
axs[0, 0].imshow(rgb_image)
axs[0, 0].set_title('RGB')
axs[0, 0].axis('off')
axs[0, 1].imshow(hsl_image)
axs[0, 1].set_title('HSL')
axs[0, 1].axis('off')
axs[1, 0].imshow(lab_image)
axs[1, 0].set_title('LAB')
axs[1, 0].axis('off')
axs[1, 1].imshow(yuv_image)
axs[1, 1].set_title('YUV')
axs[1, 1].axis('off')
plt.tight_layout()
plt.show()
```
